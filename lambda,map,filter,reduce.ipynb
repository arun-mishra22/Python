```python
## lambda fucntion 

x = lambda x : x*2

x(2)
```

```python
# map fucntion  ->  used to update every element of a iterable 

## The map() function in Python is used to apply a function to every element of an 
##   iterable (like a list, tuple, etc.) and return the results.
```

```python
L = [1,2,3,4,5]
z = map(lambda x:x**2,L)

print(z)
print(list(z))
```

```python
## filter -> used to filter the elements from a iterable

L = [1,2,3,4,5,6]

result = filter(lambda x: x % 2 == 0 , L)

print(result)
print(list(result))
```

```python
## reduce -> used to apply a fucntion to 2 value of a ietrable and reduce it to a single value . 

# It is a part of functools module . 

from functools import reduce

L = [1,2,3,4,5]

result = reduce(lambda x,y : x+y , L) 

print(result)
```
