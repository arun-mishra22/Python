## Iterators

```python
x = range(1,10)
for i in x :
    print(i)
```

```python
L = [1,2,3]
type(iter(L))

# here L is iterable but iter(L) is a iterator
```

```python
num = [1,2,3]

# fetch the iterator 
iter_num = iter(num)

# step2 --> next
next(iter_num)
next(iter_num)
```

```python
## making your own loop function 

def mera_khudka_for_loop(iterable):

    iterator = iter(iterable)

    while True :
        try:
            print(next(iterator))
        except :
            break 
```

```python
L = [1,2,3]
mera_khudka_for_loop(L)
```

```python
dir(L)  # iter only 
```

```python
dir(iter(L))  # iter and next 
```
