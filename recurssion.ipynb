```python
def factorial(x):
    if x == 1 :
        return 1 
    else :
        return x*factorial(x-1)
```

```python
factorial(5)
```

```python
# check palindrome 

def pal_check(x):
    if len(x) == 1 or len(x) == 0:
        return True 
    else :
        if x[0] == x[-1] :
            return pal_check(x[1:len(x)-1])
        else : 
            return False 
```

```python
pal_check('madam')
```

```python
## fibonaci -> 0,1,1,2,3,5,8

def fibo(x):
    if x == 1 :
        return 0 
    elif x == 2 :
        return 1 
    return fibo(x-1) + fibo(x-2)
```

```python
fibo(5)
```

```python
## memoization 

def memo_febo(x,d):
    if x in d:
        return d[x]
    else :
        d[x] = memo_febo(x-1,d) + memo_febo(x-2,d)
        return d[x]

d = {1:0,2:1} 
memo_febo(5,d)
```
